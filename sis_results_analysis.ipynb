### RNA-seq Marker Genes

```python
de_genes = pd.read_excel(os.path.join(os.getcwd(), 'data','de_genes.xlsx'), sheet_name='Table S2', header=1)
```

```python
de_genes
```

```python
de_genes_sig = de_genes[de_genes['pval'] < 3e-6]['gene_short_name']
```

```python
markers = ['NFKBIA', 'SCNN1A', 'CKB', 'PER1', 'CDH16']
```

```python
count = 0 
n = len(sis_dict['RNA Raw_class_1']['Feature'])
for ix, de_gene in enumerate(de_genes_sig):
    if de_gene in list(sis_dict['RNA Raw_class_1']['Feature']):
        print(ix, de_gene)
        count+= 1
print(f'found {count/n} DE genes')
```
